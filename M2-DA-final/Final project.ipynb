# Observations about the stats

Just at a glance, we can see there is a manufactured skew towards more of the aggressive designs, things like Offense and Mobility, while a drastically lower amount of choices that are good at Defending or Supporting

```python
df['Role'].value_counts() \
    .plot(kind= 'bar', title= 'Characters by role',)
```

When taking the roles into account, the previous scores make sense, but paint a picture of a manufactured push-focused meta, designed to help keep things moving instead of turtling around a goal ring.
